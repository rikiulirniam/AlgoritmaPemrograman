## Praktikum 04 – Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)
def biodata(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

## Praktikum 07 – Fungsi Decorator

```python
def decorator_sapa(func):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        func()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

## 7.	Fungsi dengan kwargs

## 10.	Fungsi Decorator

```python
def logger(fun):
    def wrapper():
        print("Memulai fungsi ...")
        fun()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo user, function berjalan.")

halo()
```

## 11.	Studi Kasus Nyata – Program Nilai

```python
def input_nilai():
    n = int(input("Masukkan jumlah mahasiswa: "))
    nilai = []
    for i in range(n):
        tugas = float(input(f"Nilai tugas mahasiswa ke-{i+1}: "))
        ujian = float(input(f"Nilai ujian mahasiswa ke-{i+1}: "))
        nilai.append((tugas, ujian))
    return nilai

def hitung_rata(daftar_nilai):
    total = sum([nilai for _, nilai in daftar_nilai]) / len(daftar_nilai)
    return total

def status_kelulusan(rata):
    return "Lulus" if rata >= 60 else "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

data_nilai = input_nilai()

print("\nHasil Perhitungan:")
total_rata = 0
for i, (tugas, ujian) in enumerate(data_nilai):
    akhir = nilai_akhir(tugas, ujian)
    total_rata += akhir
    print(f"Mahasiswa ke-{i+1} → Nilai Akhir: {akhir:.2f} → {status_kelulusan(akhir)}")

rata_kelas = total_rata / len(data_nilai)
print(f"\nRata-rata kelas: {rata_kelas:.2f}")
print(f"Status Kelas: {status_kelulusan(rata_kelas)}")
```
