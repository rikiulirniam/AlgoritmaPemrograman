## Praktikum 1

```python
def seq_search(data: list, target:int):
    for i in range(len(data)):
        if data[i] == target:
            print(f"Target({target}) ditemukan pada index ke {i}")
            break
    else:
        print(f"Target({target}) tidak ditemukan")

data = [10,20,30,40,50]
seq_search(data, 30)
# seq_search(data, 243)
```

## Praktikum 2

```python
def bin_search(data:list, target:int):
    start, end = 0, len(data)-1
    while start <= end:
        n = (start+end)//2
        if data[n] == target:
            print(f"Target({target}) ditemukan pada index ke {n}")
            break
        elif data[n] < target:
            start = n+1
        else:
            end = n-1
    else:
        print(f"Target({target}) tidak ditemukan")

sorted_data = [10,20,35,40,50,60,70,80,90]
bin_search(sorted_data, 40)
# bin_search(sorted_data, 209)
```

## Praktikum 5

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

## Penugasan 7

```python
import time
import pandas as pd

def sequential_search(data, target):
    for item in data:
        if item == target:
            return True
    return False

def binary_search(data, target):
    start, end = 0, len(data) - 1
    while start <= end:
        mid = (start + end) // 2
        if data[mid] == target:
            return True
        elif data[mid] < target:
            start = mid + 1
        else:
            end = mid - 1
    return False

sizes = [100, 1000, 10000]
results = []

for size in sizes:
    data = list(range(size))
    target = size - 1

    start_time = time.time()
    sequential_search(data, target)
    seq_time = time.time() - start_time

    start_time = time.time()
    binary_search(data, target)
    bin_time = time.time() - start_time

    results.append({
        "Ukuran Data": size,
        "Sequential Search (detik)": seq_time,
        "Binary Search (detik)": bin_time
    })

df = pd.DataFrame(results)
print(df)
```
