```python
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data
angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]: min_index = j
            data[i], data[min_index] = data[min_index], data[i]
    return data
angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

```python
import random, time
from bubble_sort import bubble_sort
from selection_sort import selection_sort
from insertion_sort import insertion_sort

data = [random.randint(1, 1000) for _ in range(500)]

for func in [bubble_sort, selection_sort, insertion_sort]:
    salinan = data.copy()
    start = time.time()
    func(salinan)
    end = time.time()
    print(f"{func.__name__ } membutuhkan waktu {end - start:.5f} detik")
```

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen",
10000)]

print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```
