# Praktikum

```python
%%writefile bubble_sort.py
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data
angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

```python
%%writefile selection_sort.py
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]: min_index = j
            data[i], data[min_index] = data[min_index], data[i]
    return data
angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

```python
%%writefile insertion_sort.py
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1][key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen",
10000)]

print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

```python
def desc_bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data
angka = [67,34,11,12,90,22,80]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", desc_bubble_sort(angka))
```

```python
def selection_sort(data):
    n = len(data)
    for i in range(n - 1, 0, -1):
        max_index = 0
        for j in range(1, i + 1):
            if data[j] > data[max_index]:
                max_index = j
        if max_index != i:
            data[i], data[max_index] = data[max_index], data[i]
    return data
angka = [67,34,11,12,90,22,80]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

nama_mahasiswa = ["Riki","Ulir","Niam", "Andi", "Nadine", "Yuni"]
print("Data sebelum diurutkan:", nama_mahasiswa)
print("Data setelah Insertion Sort:", insertion_sort(nama_mahasiswa))
```

```python
import random
import pandas as pd
import copy

def bubble_sort_count(arr):
    data = arr.copy()
    comparisons = 0
    n = len(data)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            comparisons += 1
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                swapped = True
        if not swapped:
            break
    return comparisons

def selection_sort_count(arr):
    data = arr.copy()
    comparisons = 0
    n = len(data)
    for i in range(n - 1, 0, -1):
        max_idx = 0
        for j in range(1, i + 1):
            comparisons += 1
            if data[j] > data[max_idx]:
                max_idx = j
        if max_idx != i:
            data[i], data[max_idx] = data[max_idx], data[i]
    return comparisons

def insertion_sort_count(arr):
    data = arr.copy()
    comparisons = 0
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            comparisons += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return comparisons

input_sizes = [10, 50, 100]
results = []

random.seed(42)

for size in input_sizes:
    test_data = [random.randint(1, 1000) for _ in range(size)]

    b_count = bubble_sort_count(test_data)
    s_count = selection_sort_count(test_data)
    i_count = insertion_sort_count(test_data)

    results.append({
        "Input Size (N)": size,
        "Bubble Sort": b_count,
        "Selection Sort": s_count,
        "Insertion Sort": i_count
    })

df = pd.DataFrame(results)
print(df.to_string(index=False))
```

```python
import os

def bersihkan_layar():
    os.system('cls' if os.name == 'nt' else 'clear')

def input_data():
    daftar_barang = []
    print("=== INPUT DATA BARANG ===")
    print("Ketik 'selesai' pada nama barang untuk berhenti input.\n")

    while True:
        nama = input("Nama Barang: ")
        if nama.lower() == 'selesai':
            break

        try:
            harga = int(input("Harga Barang: "))
            daftar_barang.append({'nama': nama, 'harga': harga})
            print(f"-> {nama} berhasil ditambahkan.\n")
        except ValueError:
            print("(!) Error: Harga harus berupa angka.\n")

    return daftar_barang

def tampilkan_tabel(data, judul="DAFTAR BARANG"):
    print(f"\n--- {judul} ---")
    if not data:
        print("Data kosong.")
        return

    print(f"{'No':<4} | {'Nama Barang':<20} | {'Harga (Rp)':>15}")
    print("-" * 45)

    for i, barang in enumerate(data, 1):
        print(f"{i:<4} | {barang['nama']:<20} | {barang['harga']:>15,}")
    print("-" * 45)

def main():
    barang = input_data()

    if not barang:
        print("Tidak ada data yang dimasukkan. Program berhenti.")
        return

    while True:
        tampilkan_tabel(barang, "POSISI DATA SAAT INI")

        print("\nPILIHAN PENGURUTAN:")
        print("1. Nama (A-Z) - Ascending")
        print("2. Nama (Z-A) - Descending")
        print("3. Harga (Termurah) - Ascending")
        print("4. Harga (Termahal) - Descending")
        print("5. Input Data Baru (Reset)")
        print("6. Keluar")

        pilihan = input("\nPilih menu (1-6): ")

        if pilihan == '1':
            barang.sort(key=lambda x: x['nama'].lower())
            print("\n>> Berhasil diurutkan berdasarkan Nama (A-Z)")

        elif pilihan == '2':
            barang.sort(key=lambda x: x['nama'].lower(), reverse=True)
            print("\n>> Berhasil diurutkan berdasarkan Nama (Z-A)")

        elif pilihan == '3':
            barang.sort(key=lambda x: x['harga'])
            print("\n>> Berhasil diurutkan berdasarkan Harga Termurah")

        elif pilihan == '4':
            barang.sort(key=lambda x: x['harga'], reverse=True)
            print("\n>> Berhasil diurutkan berdasarkan Harga Termahal")

        elif pilihan == '5':
            barang = input_data()

        elif pilihan == '6':
            print("Terima kasih!")
            break
        else:
            print("\nPilihan tidak valid, silakan coba lagi.")

if __name__ == "__main__":
    main()
```
